```python
import requests
from datetime import datetime

# Substitua pela sua chave da API
API_KEY = 'c5a93fd2e7c0bf803df2a765ccf7d689'

# Liga 71 = Campeonato Brasileiro Série A
URL = 'https://v3.football.api-sports.io/fixtures'

params = {
    'league': 71,        # ID da Série A
    'season': 2024,      # Ano desejado
    'timezone': 'America/Sao_Paulo'
}

headers = {
    'x-apisports-key': API_KEY
}

# Fazendo a requisição
response = requests.get(URL, headers=headers, params=params)
data = response.json()

# Verifica se a resposta é válida
if 'response' in data:
    for jogo in data['response']:
        rodada = jogo['league']['round']
        data_hora = jogo['fixture']['date']
        estadio = jogo['fixture']['venue']['name']
        cidade = jogo['fixture']['venue']['city']
        mandante = jogo['teams']['home']['name']
        visitante = jogo['teams']['away']['name']

        # Formatando data
        data_formatada = datetime.fromisoformat(data_hora.replace('Z', '+00:00')).strftime('%d/%m/%Y %H:%M')

        print(f"Rodada: {rodada}")
        print(f"{mandante} x {visitante}")
        print(f"Data e hora: {data_formatada}")
        print(f"Estádio: {estadio}")
        print(f"Cidade: {cidade}")
        print("="*50)
else:
    print("Erro ao obter dados da API:")
    print(data)
```
