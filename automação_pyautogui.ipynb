```python
import pynput 
from pynput.mouse import Listener as MouseListener
from pynput.keyboard import Listener as KeyboardListener, Key
import pyautogui
import time

# Lista para armazenar as coordenadas
coordenadas = []

# Função para capturar o clique do mouse
def on_click(x, y, button, pressed):
    if pressed:
        print(f"📍 Coordenadas capturadas: ({x}, {y})")
        coordenadas.append((x, y))  # Adiciona à lista

# Função para executar scroll automático
def scroll_automatico(qtd_scrolls=5, direcao='down', delay=1):
    valor_scroll = -300 if direcao == 'down' else 300
    for i in range(qtd_scrolls):
        pyautogui.scroll(valor_scroll)
        print(f"🌀 Scroll {i+1}/{qtd_scrolls} {'⬇️' if valor_scroll < 0 else '⬆️'}")
        time.sleep(delay)

# Função para capturar eventos de teclado
def on_press(key):
    try:
        # Ativa scroll automático ao pressionar "s"
        if key.char == 's':
            print("🔄 Iniciando scroll automático...")
            scroll_automatico(qtd_scrolls=5, direcao='down', delay=1)
        
    except AttributeError:
        # Pressione "ESC" para encerrar e salvar coordenadas
        if key == Key.esc:
            print("✅ Captura finalizada.")
            with open("coordenadas_salvas.txt", "w") as f:
                for coord in coordenadas:
                    f.write(f"{coord}\n")
            return False  # Encerra o listener

# Inicia os listeners
with MouseListener(on_click=on_click) as mouse_listener, KeyboardListener(on_press=on_press) as keyboard_listener:
    print("🚀 Captura iniciada.")
    print("🎯 Clique para capturar coordenadas.")
    print("🔁 Pressione 's' para scroll automático.")
    print("❌ Pressione 'ESC' para finalizar e salvar.")
    keyboard_listener.join()
```

```python
import easyocr 
import cv2
import re
import os
import csv

# 📂 Diretório de imagens e CSV
prints_path = "prints"
csv_path = "cidades_partidas_teste.csv"

# 🧠 OCR engine
reader = easyocr.Reader(['en'], gpu=False)

# 📝 CSV
with open(csv_path, mode="w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["rodada", "data", "hora", "cidade"])

    for filename in sorted(os.listdir(prints_path)):
        if filename.endswith("_full.png"):
            print(f"\n📸 Processando: {filename}")
            
            # 📥 Load da imagem
            img_path = os.path.join(prints_path, filename)
            img = cv2.imread(img_path)

            # 🔎 OCR
            results = reader.readtext(img)

            data_text, hora_text, cidade_text = None, None, None

            regex_data = re.compile(r"\d{2}[\/.-]\d{2}[\/.-]\d{4}")
            regex_hora = re.compile(r"\b\d{1,2}[:.]\d{2}\b")

            # 🧠 Detecta data e hora
            for i, (_, text, _) in enumerate(results):
                cleaned = text.replace("•", "").replace("h", ":").strip()

                if not data_text:
                    match_data = regex_data.search(cleaned)
                    if match_data:
                        data_text = match_data.group()

                        # ⏭️ Tenta pegar a hora no próximo item
                        if i + 1 < len(results):
                            prox_text = results[i + 1][1].replace("•", "").strip()
                            match_hora = regex_hora.search(prox_text)
                            if match_hora:
                                hora_text = match_hora.group().replace(".", ":")

                if not hora_text:
                    match_hora = regex_hora.search(cleaned)
                    if match_hora and not match_hora.group().startswith("20"):
                        hora_text = match_hora.group().replace(".", ":")

            # 🌆 Pega cidade após "Location"
            for idx, (_, text, _) in enumerate(results):
                if text.strip().lower() in ["location", "local", "localização"]:
                    if idx + 1 < len(results):
                        cidade_text = results[idx + 1][1].strip()
                    break

            # 🔢 Rodada do nome do arquivo
            match = re.search(r'rodada(\d+)', filename)
            rodada = match.group(1) if match else "?"

            writer.writerow([rodada, data_text or "?", hora_text or "?", cidade_text or "?"])
            print(f"✅ RODADA {rodada} | 📅 {data_text} | ⏰ {hora_text} | 🌆 {cidade_text}")
```

```python
import easyocr  
import cv2
import re
import os
import csv

# 📂 Diretório de imagens e CSV
prints_path = "prints"
csv_path = "cidades_partidas_teste.csv"

# 🧠 OCR engine
reader = easyocr.Reader(['en'], gpu=False)

# 📝 CSV
with open(csv_path, mode="w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["rodada", "data", "hora", "cidade"])

    for filename in sorted(os.listdir(prints_path)):
        if filename.endswith("_full.png"):
            print(f"\n📸 Processando: {filename}")
            
            # 📥 Load da imagem
            img_path = os.path.join(prints_path, filename)
            img = cv2.imread(img_path)

            # 🔎 OCR
            results = reader.readtext(img)

            data_text, hora_text, cidade_text = None, None, None

            regex_data = re.compile(r"\d{2}[\/.-]\d{2}[\/.-]\d{4}")
            regex_hora = re.compile(r"\b\d{1,2}[:.]\d{2}\b")

            # 🧠 Detecta data e hora
            for i, (_, text, _) in enumerate(results):
                cleaned = text.replace("•", "").replace("h", ":").strip()

                if not data_text:
                    match_data = regex_data.search(cleaned)
                    if match_data:
                        data_text = match_data.group()

                        # ⏭️ Tenta pegar a hora no próximo item
                        if i + 1 < len(results):
                            prox_text = results[i + 1][1].replace("•", "").strip()
                            match_hora = regex_hora.search(prox_text)
                            if match_hora:
                                hora = match_hora.group().replace(".", ":")
                                h, m = map(int, hora.split(":"))
                                if 6 <= h <= 23:
                                    hora_text = hora

                if not hora_text:
                    match_hora = regex_hora.search(cleaned)
                    if match_hora:
                        hora = match_hora.group().replace(".", ":")
                        try:
                            h, m = map(int, hora.split(":"))
                            if 6 <= h <= 23:
                                hora_text = hora
                        except:
                            continue

            # 🌆 Pega cidade após "Location"
            for idx, (_, text, _) in enumerate(results):
                if text.strip().lower() in ["location", "local", "localização"]:
                    if idx + 1 < len(results):
                        cidade_text = results[idx + 1][1].strip()
                    break

            # 🔢 Rodada do nome do arquivo
            match = re.search(r'rodada(\d+)', filename)
            rodada = match.group(1) if match else "?"

            writer.writerow([rodada, data_text or "?", hora_text or "?", cidade_text or "?"])
            print(f"✅ RODADA {rodada} | 📅 {data_text} | ⏰ {hora_text} | 🌆 {cidade_text}")
```

```python
import csv

input_file = "cidades_partidas_teste.csv"       # arquivo original
output_file = "cidades_partidas_teste_sorted.csv" # arquivo com as rodadas ordenadas

with open(input_file, newline='', encoding="utf-8") as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)  # lê o cabeçalho
    rows = list(reader)

# Ordena as linhas convertendo a rodada para inteiro (caso sejam números)
rows_sorted = sorted(rows, key=lambda r: int(r[0]) if r[0].isdigit() else r[0])

with open(output_file, "w", newline='', encoding="utf-8") as outfile:
    writer = csv.writer(outfile)
    writer.writerow(header)
    writer.writerows(rows_sorted)

print(f"Arquivo '{output_file}' criado com as rodadas ordenadas.")
```

```python
import csv

# Função para processar o arquivo CSV
def processar_csv(arquivo_entrada, arquivo_saida):
    with open(arquivo_entrada, 'r', encoding='utf-8') as entrada, \
         open(arquivo_saida, 'w', newline='', encoding='utf-8') as saida:
        
        # Criar leitor e escritor CSV
        leitor = csv.reader(entrada)
        escritor = csv.writer(saida)
        
        # Escrever o cabeçalho sem a coluna "rodada"
        cabecalho = next(leitor)
        novo_cabecalho = [col for col in cabecalho if col.lower() != 'rodada' and col.lower() != 'rodadae']
        escritor.writerow(novo_cabecalho)
        
        # Processar cada linha
        for linha in leitor:
            # Ignorar linhas vazias
            if not linha:
                continue
            
            # Mapear colunas com seus valores
            valores = {}
            for i, col in enumerate(cabecalho):
                if i < len(linha):
                    valores[col.lower()] = linha[i]
            
            # Processar o campo cidade para remover ", Brazil"
            cidade = valores.get('cidade', '')
            if cidade.endswith('; Brazil'):
                cidade = cidade[:-8]  # Remove "; Brazil"
            elif cidade.endswith(', Brazil'):
                cidade = cidade[:-8]  # Remove ", Brazil"
            
            # Criar nova linha sem o campo "rodada"
            nova_linha = []
            for col in cabecalho:
                if col.lower() != 'rodada' and col.lower() != 'rodadae':
                    if col.lower() == 'cidade':
                        nova_linha.append(cidade)
                    else:
                        nova_linha.append(valores.get(col.lower(), ''))
            
            # Escrever a nova linha
            escritor.writerow(nova_linha)

# Nome dos arquivos
arquivo_entrada = 'cidades_partidas_teste_sorted.csv'  # Nome do seu arquivo de entrada
arquivo_saida = 'cidades_partidas_filtrado.csv'  # Nome do arquivo de saída

# Executar o processamento
processar_csv(arquivo_entrada, arquivo_saida)
print(f"Arquivo processado com sucesso! Resultado salvo em 'cidades_partidas_filtrado.csv'")
```
