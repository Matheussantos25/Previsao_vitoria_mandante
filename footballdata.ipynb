```python
import pandas as pd

# Carregar os arquivos
df_2024 = pd.read_csv("df_2024.csv")
df_cidades = pd.read_csv("cidades_partidas_teste_sorted.csv")

# Padronizar 'data'
df_2024['data'] = pd.to_datetime(df_2024['data'])
df_cidades['data'] = pd.to_datetime(df_cidades['data'])

# Tratar 'Time' no df_2024
df_2024['Time'] = pd.to_datetime(df_2024['Time'], format='%H:%M', errors='coerce').dt.time

# Remover valores inválidos de hora no df_cidades
df_cidades = df_cidades[df_cidades['hora'] != '?']

# Tratar 'hora' no df_cidades
df_cidades['hora'] = pd.to_datetime(df_cidades['hora'], format='%H:%M', errors='coerce').dt.time

# Realizar merge pelas colunas data + hora
df_merged = pd.merge(
    df_2024,
    df_cidades[['data', 'hora', 'rodada', 'cidade']],
    left_on=['data', 'Time'],
    right_on=['data', 'hora'],
    how='left'
)

# Limpar coluna duplicada
df_merged.drop(columns=['hora'], inplace=True)
df_merged
```
